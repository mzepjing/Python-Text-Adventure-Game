```python
# Import sleep for small pause during the game
# Import pyfiglet to use ASCII art form of text
# Import Timer for timing the user's action
# Import keyboard for cancelling the user input
# Import TextBlob for sentiment analysis
# Import random for later use
from time import sleep
import pyfiglet
from threading import Timer
import keyboard
from textblob import TextBlob
import random






##########################################
# Lose function for the game
def lose():
    
    # Set up
    dead_man = '''
--------------------------------------
You were killed!!!
                                                                                  ...:...                               
                                                                    ........    ::;..;+.::.                             
                                                             ....:;;;.......:..:;;:.:;;. :.                             
                                                  ..........::;;;::.             .;;;::.::                              
                                                :;;;:::::::::::::.               .; ...::.                              
                        .::.             .......::::::....     .:;:         .:;:  .:                                    
                        ;. :...................                   :;     .:;::. ..:.                                    
                        ::  .........         ........::          .+:...:;:.  .::.                                      
                         :.  .......................:::.        ..:.  .:.   .::.                                        
                          ............           .:..      .......  .:;:...:..                                          
                                               .:.     .......   .::...;:..                                             
                                             .::.   ..:..        ;+;::::                                                
                                           :;:.   .::.           .:;;;.                                                 
                                          :+...:.:..                                                                    
                                       ..;; ..;;..                                                                      
........................................................................................................................
'''
        
    # Print the result for player's action
    print(dead_man)
    sleep(1)
    print(pyfiglet.figlet_format('''Game Over!!'''))
    
    # If the player lose, ask for replay
    # Give the player 5 attempts, afterwards end the game
    for attempt in range(5):
        
        # Ask user input for replay
        action = input(prompt = '''Would you like to restart the game?(Yes/No)''').lower()

        # If the answer is yes then restart the game
        if 'y' in action:
            start_game()
            break

        # If the answer the no then end
        elif 'n' in action:
            break

        # If the input is wrong then ask again
        else:
            print('\n    Please enter again.\n')
##########################################


    
##########################################
# Win function for the game
def win():
    
    # Set up
    fireworks = '''
--------------------------------------
You escape the prison!!!
                                             ,                    ,                     ,,#@@                                     
                                                #@@              +:               :o@@*                                           
                                                   +@#          o            :@+                                                  
                                           ,+         @+       @         @@                                                       
                                                +s,    ,@      ,      sz     :@o;,                                                
                                                    *+   +,        ;*   o                   ;                                     
                               *@@o,@@@@@@@@@z,       ,   s,      :  :         *s@@@@@@soz, ::,                                   
                                                    ;@     +          z#,                                                         
                                                                       ,+zzo@@@o*:                                                
                                     #;@              ,o:            :#o,                                                         
                                                 +@,     ,    :             ;#s                                                   
                                             @z       o      o+      o            @@,                                             
                                          @@,       @    o   @,        +             @@*                                          
                                       #@+        ,z    ,;  ,@,          z              @@s                                       
                                     @@          #o    ,@   ,@*          ,:               ,;;                                     
                                   ,@*                 #@    @@                            @@@                                    
                                                       @@    ,:                                                                   
                                                      ;@@     +,,                                                                 
                                                      z@@                                                                         
        *:                                            o@@                                          ,@,                      ,     
          ,*   ,    zz                                z@o                                             z@      ,        zz:        
          *  , + + +:                                 z@z                                          ,,    #    ,    ;,             
           ;*    *,                                   ;@z                                               z  ; ,  ; +*         ,    
         ,; , ;:;   ;@   *                             @s                                               *  ,     ++,              
      ,+  :   s         o,     @z,                     @@                                        ++      ,;    ;+;       :        
     ,        o*                    ;@@;               s@                                 ;@@*      ,s   +   s   ,     z+         
                                        ,z@            ,@                          +@@s           @    ,    z*    +;      *o      
                                             :@+        ,z                    @z,              ;,     ;:    z@     ,+        z,   
                                                 @,      @                +s                                 *;                   
                                                    #     *            s                                                          
                                                      #   ;o         s                                                            
                                                        ;   ;      :                                                              
.........................................................,........,...............................................................
'''
        
    # Print the result for player's action
    print(fireworks)
    sleep(1)
    print(pyfiglet.figlet_format('''Congratulations!!'''))
##########################################
    
    
    
    
    

##########################################
# Start of the game
def start_game():
    
    # Set up
    intro = '''
Story of the game:

Dale Conti, a college student, was on his way to become a paramedic.
But, he fell for heroin. His life became a mess. Months ago, he robbed a bank.
His criminal partner killed a cop, and ran away, but Dale got caught.

Wally Nightingale, the cop who's in charge of Dale's case,
is desperate to find out Dale’s criminal partner, the killer of the cop.
Wally’s obsession with catching murderers turns him into an unscrupulous policeman.
He elaborately spread the rumor, which Dale is a snitcher.

Now, everyone in the prison thinks Dale is a dog (a snitcher).
Everyone want him to be dead. Especially, Barnes. He is planning to kill Dale.


Description of the game:

You are playing as Dale Conti. Your goal is to escape the prison.
Your "best" friend Ned might have a plan. You should talk to him.
Be careful with you action in the prison!!! You might die.

In Stage #2, You can explore the room, but it will not effect the game if you don't.


Be aware:

TextBlob is not perfect at identifying the polarity value of the text.
There are so many ways to cheat TextBlob as well.
In stage #2 when ask "Say something nice to Ned."
The input best to be straight forward compliment.
'''
    
    # Direct quote from the show
    # Reference:
    # Singer, E. (Writer), Lightfoot, S. (Writer) & Nalluri, B. (Director). (2022, October 14). 
    # The Bombay No (Season 1, Episode 1) [TV series episode]. 
    # In S. Lightfoot, A. Barron, B. Nalluri, N. Clemens, S. Golin, E. W. Singer, J. Kurzel, D. Manson, 
    # R. Sharkey (Executive Producer), Shantaram. Fair Honest Positive Creative; The 4 Keys; Bohemian Risk Productions;
    # Square Head Productions; Anonymous Content; Paramount Television Studios.
    quote = '''
"It took me a long time and some hard lessons to learn what I know about love, fate and chasing redemption.
That story saw me beaten, loved, befriended and chained on three continents.
But it started with a simple choice escape or die." (Singer et al., 2022)
'''
    
    # Print the description of the game
    print(intro)
    
    input(prompt = '    <Press Enter to start game>').lower()
    
    print(quote)

    # Present the title screen
    print(pyfiglet.figlet_format('''Shantaram'''))
    
    easter_egg_timer()
##########################################
    
    

##########################################
# User defined function when time is out, execute the function
def easter_egg():
    
    # Set up
    easter_egg = '''
Congratulations!!!
You find the easter egg.

                ********                
             z************z             
           ******************           
         :********************,         
        *******:********+::*****        
       *******:::+*****:::::*****       
     :::*****::::::***:::::::****:      
    ::::::*:::::::::+:::::::::;*:::     
    **:::::::::**::::::::**:::::::::    
   *****::::::****::::::****:::::::**   
  *******;:::*******:::******:::::***z  
  *********+*********:*********::*****  
 ************************************** 
;************************************** 
****************************************
*****z************z***********z*********
****zzzz********zzzz*********zzzz*******
z*zzzzzzz******zzzzzzz*****zzzzzzz*****z
 zzzz*zzzzz**zzzzz*zzzz**zzzzzzzzzzz**z;
  *****zzzzzzzzz****zzzzzzzzz***zzzzzzz 
   ******zzzzzz*******zzzzz*******zzz,  
    :******zz**********zzz**********    
      ,***************************      
          ********************z         
              ************,             
........................................

Did you know that, in the game of Far Cry 4, 
you can trigger an alternative ending by waiting for 15 mins after the start of the game.
'''
    
    # Cancel the user input box automatically then print the easter egg
    keyboard.press_and_release('enter')
    print(easter_egg)
##########################################

    

##########################################
# Set up for an easter egg
def easter_egg_timer():
        
    # Set up the timer for 15 second, if no input then print the easter egg
    timeout = 15
    t = Timer(timeout, easter_egg)
    
    # Start the timer function
    t.start()

    print('\n    Loading an egg', end='')
    
    # For loop to print '.' every second
    for i in range(5):
        print('', end='.')
        sleep(1)

    # Give the user choice to skip the easter egg
    input(prompt = '\n    <Press Enter to SKIP the egg>').lower()
    
    # End the timer
    t.cancel()
    
    prison_playground()
##########################################






##########################################
# Stage #1: Prison playground
def prison_playground():
    
    # Set up
    stage1 = '''
--------------------------------------
Part 1 - A back stabber

Ned: "Watch out, Dale, behind you!"

    Choose your action:
    1. Who's Dale?
    2. "What's going on, Ned?"
    3. Look behind.
'''
    
    # Result for each action inputs
    action_result1 = '''
You still wondering this question?!

You name is Dale Conti. You were a college student, on your way to become a paramedic. 
But, you fell for heroin. Your life became a mess. Months ago, you robbed a bank. 
Your criminal partner killed a cop. He ran away, but you got caught.
'''
    action_result2 = '''
It's a really "great" time for chat! 
Ned doesn't want to respond. He thinks you are already a dead man.
'''
    action_result3 = '''
Now, Barnes rushed at you from behind. He stabbed you with a sharpened tooth brush.
You couldn't react to that, because you are still wondering what is going on.
'''
    
    # Death warning for each actions
    death1 = '''
Now, Barnes rushed at you from behind. He stabbed you with a sharpened tooth brush.
You couldn't react to that, because you are still wondering who you are. '''
    death2 = '''
Now, Barnes rushed at you from behind. He stabbed you with a sharpened tooth brush.
You couldn't react to that, because you are still wondering what is going on.'''
        
    print(stage1)
    
    while True:
        
        # Ask for input then check if the input is valid
        action = input(prompt = '    Action: ').lower()

        # If the player's choice is 1, give the result of the action
        if 'dale' in action or '1' in action:
            
            print(action_result1)

            # Pause for the player to read the text
            print(death1)
            
            # Player get kill, game over
            lose()
            break

        # If the player's choice is 2, give the result of the action
        elif 'ned' in action or '2' in action:
            
            print(action_result2)

            # Pause for the player to read the text
            input(prompt = '    <Continue>').lower()
            print(death2)

            # Player get kill, game over
            lose()
            break

        # If the player's choice is 3, give the result of the action
        elif 'behind' in action or '3' in action:
            
            print(action_result3)

            # Ask the player to input for further interaction
            action = input(prompt = '    Enter "L" to dodge: ')

            # If the player enter the correct input, give the result of the action
            if action in ['L', 'l']:
                
                # Give the player 80% chance to successfully dodge the attack
                dodge = random.randint(0, 5)

                # If the dodge = 0, meaning the player didn't successfully dodge the attack
                # Then the player will lose the game
                if dodge == 0:

                    # Print the text for player to understand what happened
                    print('\n    You tried to dodge, but failed.')

                    # Player lose the game. Break the loop
                    lose()
                    break

                # If dodge != 0 then continue the game
                else:
                    pass
                
                print('''\nYou dodged the attack! That's great!\n ''')

                # Ask for the input, then use the input to generate a text
                body_part1 = input(prompt = '    Enter a body part: ').lower()
                body_part2 = input(prompt = '    Enter another body part: ').lower()
                body_part3 = input(prompt = '    Enter another body part: ').lower()
                cute_animal = input(prompt = '    Enter a cute animal name: ').lower()
                
                # Print the result for player
                print(f'''
You grabbed his {body_part1}, pulled him over, and punched his {body_part2} with your fist, 
then you kicked his {body_part3}. He was hurt, and made a {cute_animal} noice.

Wow!! You are really good at fighting!

This should teach him a lesson.
Barnes will not bother you for a while, but you need to escape the prison ASAP.''')
                
                prison_cell()
                break

            # If the player didn't enter the correct input, give the result of the action
            else:
                
                print('''\nYou didn't react to that. Barnes stabbed you. ''')
                
                # Player get kill, game over
                lose()
                break
                
        # If the input is not the number of the action, go back to first action input
        else:
            
            print('''\n    Sorry, I didn't understand. Please type again.\n''')
##########################################






##########################################
# Stage #2: Prison cell
def prison_cell():
    
    # Set up
    stage2 = '''
--------------------------------------
Part 2 - Praise him hard

You are now back to your prison cell.
Ned walks in shortly after.'''
    
    # Text for the inputs to interact
    action1 = '''
    What do you want to do?
    [Check the room] (Not necessary), [Talk to Ned]
'''
    action2 = '''
    Choose your action
    [Bed], [Desk], [Mirror], [Go back to last question]
'''
    
    # The objects that can be interact with
    bed = '''
You find a spoon under the bed.

Ned: "What the hell is the spoon for?
      Are you going to escape the prison with that?
      You know what. I'll keep it for you."
'''
    desk1 = '''
There are some books you have been reading.
'''
    desk2 = '''
Ned: "'Misery acquaints a man with strange bedfellows.' (Shakespeare, 2022)
      I heard that the warden loves Shakespeare.
      If there's a safe, I bet the combination is his birthday."
'''
    mirror = '''
The scars are all over you body.

Ned: "You do know that this will not be their last attempt to kill you, right?"
'''
    
    # Talking to Ned options set up
    talk_ned = '''
Dale: "I must get out of here. 
       They are going to kill me."
Ned: "Just tell the cop what he wants to know.
      He will get you out of here."
Dale: "What?
       I'm NOT A FXXXING DOG!
       You need to help me."
       
    Say something nice to Ned. He might tell you his plan to escape.
'''
    talk_tip = '''    --------------------------------
    Try to type: You are a good man.
    --------------------------------
    '''
    talk_ned_f = '''
Ned: "I can't help you."

You are on you own.
Shortly after, the prisoners come to you.
This time you didn't have the chance to fight back. '''
    
    # Ned response with your words
    ned_plan1 = '''
Ned: "I do have a plan.
      You know that the workshop is just on the edge of this prison.
      If we can get to the rooftop, we might have a chance to escape."
'''
    ned_plan2 = '''
Ned: "What the hell are you doing on my back?
      I do have a plan though.
      You know that the workshop is just on the edge of this prison.
      If we can get to the rooftop, we might have a chance to escape."
'''
    
    # Set a list for books for later use
    books = ['Les Misérables', 'The Godfather', 'The Tempest']
    
    input(prompt = '\n    <Continue>').lower()
    
    # Print the title of the stage 2
    print(stage2)
    
    # Loop for user to explore the room
    while True:
        
        # Print the first action that the player can choose from
        print(action1)
        
        action = input(prompt = '    Type your action: ').lower()
        
        # If the user wants to check the room
        if action in ['c', 'check', 'check the room']:
            
            # Loop for user to explore the room
            while True:
            
                print(action2)
                
                # Action input set to lower case
                action = input(prompt = '    Interact with the room: ').lower()
                
                # If the user wants to interact with bed, print the result
                if action in ['b', 'bed']:
                    
                    # Give player 50% chance to trigger an easter egg
                    spoon_find = random.randint(0, 1)
                    
                    # If the spoon_find = 1, then print the easter egg
                    if spoon_find == 1:
                        print(bed)
                    
                    # If the spoon_find = 0, then print there's nothing
                    else:
                        print('There is nothing here.')
                    
                # If the user wants to interact with desk, print the result
                elif action in ['d', 'desk']:
                    print(desk1)
                    
                    # Print the books
                    for book in books:
                        print(f'''{book}''')
                    
                    # Reference:
                    # Shakespeare, W. (2022). The Tempest. Independently published.
                    print(desk2)
                    
                # If the user wants to interact with mirror, print the result
                elif action in ['m', 'mirror']:
                    print(mirror)
                
                # If the user wants to go back, go back to the last loop
                elif action in ['g', 'back', 'go back', 'go back to last question']:
                    break
                    
                # If the input is not vaild, back to the beginning of the loop
                else:
                    print('\n    Please type your input again')
                
        # If the uer wants to interact with Ned, print the result
        elif action in ['t', 'talk', 'ned', 'talk to ned']:
            print(talk_ned)
            
            # Set the attempt to 1. If the user can't give the right answer, then lost the game
            attempt = 1
            while attempt <= 3:
                
                # When the user is on 4th attempt, give out the tip for the answer
                if attempt == 3:
                    print(talk_tip)
                
                # Else then pass, the game continue
                else:
                    pass
                
                # Ask for input to be sentences
                action = input(prompt = '    Praise him hard with sentences: ').lower()

                # TextBlob to determine if the input is positive
                # If the positive, means player is accomplishments Ned
                if TextBlob(action).sentiment.polarity > 0:
                    print(ned_plan1)
                    
                    # If get Ned's plan for escaping the prison, then go to next stage
                    workshop()
                    break
                    
                # If the player type to kiss Ned's ass, print the result
                elif action in ['kiss', 'mua', 'kiss his ass']:
                    print(ned_plan2)
                    
                    # If get Ned's plan for escaping the prison, then go to next stage
                    workshop()
                    break
                
                # TextBlob to determine if the input is negative
                # If the negative, means player is not compliment Ned
                elif TextBlob(action).sentiment.polarity < 0:
                    
                    # The attempt increment by 1
                    attempt += 1
                    
                    # Check if the attempt is 5. Yes, then the player lost the game
                    if attempt == 4:
                        print(talk_ned_f)
                        lose()
                        break
                    
                    # Check if the attempt is 5. No, then continue
                    else:
                        print('''\n    Do you want his help or not?\n''')
                    
                # If the sentiment analysis get 0, means neither positive input receive nor negative input receive
                # Then ask the player try again
                else:
                    
                    # The attempt increment by 1
                    attempt += 1
                    
                    # Check if the attempt is 5. Yes, then the player lost the game
                    if attempt == 4:
                        print(talk_ned_f)
                        lose()
                        break
                        
                    # Check if the attempt is 5. No, then continue
                    else:    
                        print('''\n    Just kiss his ass!!\n''')

            # Break when the player failed to know Ned's plan for escaping
            break

        # If the input for choose object to interact is wrong, ask for input again
        else:            
            print('\n    Please type you input again')
##########################################






##########################################
# Stage #3: Workshop
def workshop():
    
    # Set up
    stage3 = '''
--------------------------------------
Part 3 - Escape or die

It's lunch time. The renovation worker in the workshop are gone for now.
You and Ned sneak to the door.
But, the door is locked.

Look like a 4 digit combination.'''
    
    # Actions with the related outputs
    action_result1 = '''
You punched the lock so hard it almost break you hand. 
Try something better.'''
    action_result2 = '''
Dale: "Hey, Ned.
       We got a situation here.
       Any chance you would know the combination?"
Ned: "How would I know?
      ...
      Well, maybe try 0423."'''
    
    # Conversation once player in workshop
    in_workshop = '''Now, you are one step closer to freedom.
Find a way to get to the rooftop.

Ned: "Now, go find a ladder, a long cable, and something to break the celling."

    Here are the items in the workshop.
'''
    
    # Set up for the workshop
    workshop = ['Duct tape', 'Hammer', 'Crowbar', 'Sponge', 'Wrench',\
                'Screwdriver', 'Utility knife', 'Ladder', 'Handsaw', 'Long cable']
    
    # Set the necessary item to False, meaning dale doesn't have the item yet 
    long_cable = False
    ladder = False
    Dale_invent = []
    
    input(prompt = '\n    <Continue>').lower()
    
    # Print the title of the stage 3
    print(stage3)
    
    # Loop for choosing the action
    while True:
        
        action = input(prompt = '''
    Choose you action:
    
    1. Try to break the lock
    2. Talk to Ned
    3. Input the combination

    Action: ''').lower()
        
        # If the player's choice is 1, give the result of the action
        if '1' in action or 'break' in action:
            print(action_result1)
        
        # If the player's choice is 2, give the result of the action
        elif '2' in action or 'ned' in action or 'talk' in action:
            print(action_result2)
        
        # If the player's choice is 3, give the result of the action
        elif '3' in action or 'input' in action or 'combination' in action:
            
            # Check if the user input is correct.
            while True:
                
                combination = input(prompt = '\n    Type the combination: ').lower()
                
                # Check if the user input is correct.
                try:

                    # Check if the input is 4 digits. 1 is same as 0001
                    if 0 <= int(combination) <= 9999:
                        
                        # If the combination is correct, player can enter the room
                        if int(combination) == 423:
                            
                            # Set combination_is_wrong to False when it's right
                            combination_is_wrong = False
                            print('\nYou unlock the door.')
                            break
                        
                        # If the combination is wrong, ask the input again
                        else:
                            
                            # Set combination_is_wrong to True
                            combination_is_wrong = True
                            break
                    
                    # If the input is not the number of 4 digits, ask the user to input again.
                    else:
                        print('\n    4 digit at max!')

                # Except when the user input is anything else but an integer (string, or float, or empty value...).
                except:
                    print('\n    Please enter a number.')
            
            # If the combination is wrong go back to action 1
            if combination_is_wrong == True:
                print('\n    Wrong combination! Please try a different action.')
            
            # If the combination is right then break the first loop
            else:
                break
                        
        # If the input is not correct, ask for input again
        else:
            print('\n    Please enter a number.\n')
    
    # Give the list of the item once the player enter the workshop
    print(in_workshop)
        
    for item in workshop:
        print(item)
    
    print('')
    
    # Loop for player to choose the item
    while True:
        
        item = input(prompt = '    Type the item you want to take: ').capitalize()
        
        # Check if the item match the item in workshop list
        if item in workshop:
            
            # The right item goes into dale inventory
            Dale_invent.append(item)
            
            # Check if the player get ladder
            if 'Ladder' in item:
                ladder = True
                
                # If player get all items needed then continue the game
                if ladder == True and long_cable == True and len(Dale_invent) >= 3:
                    break
                    
                # Else then pass, the game continue
                else:
                    pass
                
            # Check if the player get long cable
            elif 'Long cable' in item:
                long_cable = True
                
                # If player get all items needed then continue the game
                if ladder == True and long_cable == True and len(Dale_invent) >= 3:
                    break
                
                # Else then pass, the game continue
                else:
                    pass
                    
            # If player get all items needed then continue the game
            elif ladder == True and long_cable == True and len(Dale_invent) >= 3:
                break
            
            # Else then pass, the game continue
            else:
                pass
          
        # If the item is not the workshop list, ask for input again
        else:
            print('\n    Type the right name.\n')
    
    # Keep only items that is not ladder and long cable from dale inventory
    Dale_invent.remove('Ladder')
    Dale_invent.remove('Long cable')
    
    # Find a random item in dale inventory
    index = random.randint(1, len(Dale_invent))
    
    print(f'''
You got what you need.
Ned climb the ladder you took, and use {Dale_invent[index-1].lower()} to break the ceiling.
You and Ned both get to the rooftop.
With the long cable you took, you were both able to climb over the prison wall.
''')
    
    win()
##########################################

start_game()
```
